```python
import pandas as pd
import numpy as np
```

```python
#Read the three files into python as dataframes
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
file1
```

```python
#Standardize header names & rearrange
file1 = file1.rename(columns = {'ST':'State',
                               'GENDER':'Gender'})
cols1 = []
for c in file1.columns:
    cols1.append(c.lower())
file1.columns = cols1

file1 = file1[['customer','gender','state','education','income','customer lifetime value','vehicle class','monthly premium auto','policy type','number of open complaints','total claim amount']]

file1
```

```python
file2
```

```python
#Standardize header names & rearrange
file2 = file2.rename(columns = {'ST':'State',
                               'GENDER':'Gender'})
cols2 = []
for c in file2.columns:
    cols2.append(c.lower())
file2.columns = cols2

file2 = file2[['customer','gender','state','education','income','customer lifetime value','vehicle class','monthly premium auto','policy type','number of open complaints','total claim amount']]

file2
```

```python
file3
```

```python
#Standardize header names & rearrange.
cols3 = []
for c in file3.columns:
    cols3.append(c.lower())
file3.columns = cols3

file3 = file3[['customer','gender','state','education','income','customer lifetime value','vehicle class','monthly premium auto','policy type','number of open complaints','total claim amount']]

file3
```

```python
#Concatenate the three dataframes
merged_data = pd.concat([file1, file2, file3], axis=0)
merged_data
```

```python
#Which columns are numerical & categorical?
merged_data.dtypes
```

```python
merged_data._get_numeric_data()
```

```python
#Delete the column education and the number of open complaints from the dataframe.
merged_data = merged_data.drop(['education','number of open complaints'], axis=1)
merged_data
```

```python
#Correct the values in the column customer lifetime value. 
merged_data['customer lifetime value'] =  pd.to_numeric(merged_data['customer lifetime value'], errors='coerce')
merged_data
```

```python
#Correct the values in the column customer lifetime value. 
merged_data['customer lifetime value'] = merged_data['customer lifetime value'] / 100
merged_data
```

```python
#Check for duplicate rows in the data and remove if any.
merged_data = merged_data.drop_duplicates()
merged_data
```

```python
#Filter out the data for customers who have an income of 0 or less.
merged_data[merged_data['income']>0.0]
```
